```python
import numpy as np
students = np.random.randint(50, high=101, size=(50, 5))
students.shape
```

```python
# ndarray的實體方法
students.sum(axis=1)
students.mean(axis=1)
```

```python
# numpy內建的函式

np.sum(students, axis=1)
np.mean(students, axis=1)
```

```python
# 1家分站30天的營收

oneStore = np.random.randint(100, 10000, size=(30,))
oneStore
```

```python
# 累計每天這家店的營收
import matplotlib.pyplot as plt

oneStoreCum = oneStore.cumsum()
oneStoreCum

figure = plt.figure(figsize=(10, 5))
axes = figure.add_subplot(1,1,1)
axes.plot(range(1, 31), oneStoreCum, 'r-')
axes.set_title("one Store shop cumsum")
axes.set_xlabel("day")
axes.set_ylabel("money")
plt.show()
```
