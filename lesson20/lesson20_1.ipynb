```python
import pandas as pd

# 美國球賽的小費數據集
tips = pd.read_csv('tips.csv')
tips.info()
```

```python
# 改變tips的欄位名稱
tips.columns = ['總票價', '小費', '是否吸煙', '日期', '時間', '人數']
tips
```

```python
# 重新排序欄位的順序
tips = tips[['日期', '時間', '是否吸煙', '人數', '總票價', '小費']]

# 新增一個欄位:小費比列
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
# 依照日期,建立一個群組
grouped = tips.groupby('日期')
len(grouped)
```
