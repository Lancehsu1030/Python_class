```python
import numpy as np
import matplotlib.pyplot as plt

# 指定間隔數量,傳出固定間隔的數字
x = np.linspace(-3, 3, 50)

#以下是陣列運算
y1 = 2 * x + 1
y2 = x ** 2

plt.plot(x, y1, color='red', label='y=2x+1', linewidth=2, linestyle='--')
plt.plot(x, y2, label='y=x^2', linewidth=2, color='blue', linestyle='-')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Two Functions')
plt.legend()
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.grid()
plt.show()
```
