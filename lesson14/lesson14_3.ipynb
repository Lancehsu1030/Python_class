```python
import numpy as np
import matplotlib.pyplot as plt

# 指定間隔數量,傳出固定間隔的數字
x = np.linspace(-3, 3, 50)

#以下是陣列運算
y1 = 2 * x + 1
y2 = x ** 2

plt.plot(x, y1, color='red', label='y=2x+1', linewidth=2, linestyle='--')
plt.plot(x, y2, label='y=x^2', linewidth=2, color='blue', linestyle='-')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Two Functions')
plt.legend()
plt.xlim((-4, 4))
plt.ylim((-1, 5))
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 3, 0.02)
# 正弦波
y1 = np.sin(2 * np.pi * t)
# 指數衰減
# 請使用繁體中文
y2 = np.exp(-t)
plt.plot(t, y1, 'bo--', label='sin')
plt.plot(t, y2, 'rs--', label='exp')
plt.xlabel('時間 (秒)')
plt.ylabel('振幅')
plt.title('阻尼正弦波')
plt.legend()
plt.grid()
plt.show()
```
