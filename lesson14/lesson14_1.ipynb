```python
import matplotlib.pyplot as plt

student1 = {"國文":78, "英文":80, "數學":92, "歷史":75, "探索":85}
student2 = {"國文":92, "英文":60, "數學":85, "歷史":62, "探索":73}

# 畫線圖, x軸為科目, y軸為分數
subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

#
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,scores1)
axes.plot(subjects,scores2)
plt.show()
```

```python
import matplotlib.pyplot as plt

student1 = {"國文":78, "英文":80, "數學":92, "歷史":75, "探索":85}
student2 = {"國文":92, "英文":60, "數學":85, "歷史":62, "探索":73}

subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

figure = plt.figure(dpi=100, figsize=(8,5), facecolor='lightgray')
axes = figure.add_subplot(1,1,1)
axes.set_title("學生分數比較")
axes.set_xlabel("科目")
axes.set_ylabel("分數")

# 在 plot 時加上 label，然後呼叫 legend()
# 自訂line的style

axes.plot(subjects, scores1, marker='o', label='學生1', linestyle='--')
axes.plot(subjects, scores2, marker='s', label='學生2', linestyle='-.')
axes.legend()

plt.show()
```
