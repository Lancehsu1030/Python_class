```python
class Person:
    def __init__(self,name:str,age:int): #自訂的初始化
        self.name = name
        self.age = age
```

```python
p1 = Person("robert_hsu",30)
print(p1.name)
print(p1.age)
```

```python
p2 = Person(age=35,name="alice")
print(p2.name)
print(p2.age)
```
